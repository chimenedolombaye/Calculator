```python
import math
# Créez une classe appelée « Calculatrice » qui contient les éléments suivants :
class calculatrice:
#Un attribut de dictionnaire pour stocker les opérations mathématiques disponibles et leurs fonctions correspondantes
#Une méthode appelée "init" qui initialise le dictionnaire avec les opérations mathématiques de base (+, -, *, /) et les fonctions correspondantes
    def __init__(self):
        self.operations = {
            '+': self.addition,
            '-': self.soustraction,
            '*': self.multiplication,
            '/': self.division,
        }
# Une méthode appelée "add_operation" qui prend deux arguments : le symbole de l'opération et la fonction correspondante.
# Cette méthode doit ajouter la nouvelle opération et la nouvelle fonction au dictionnaire.

    def addition(self, a, b):
        return a + b

    def soustraction(self, a, b):
        return a - b

    def multiplication(self, a, b):
        return a * b

    def division(self, a, b):
        if b == 0:
            return "Erreur : Division par zéro."
        else:
            return a / b

    def add_operation(self, symbole, fonction):
        self.operations[symbole] = fonction

    def calculer(self, a, symbole, b):
        if symbole not in self.operations:
            raise ValueError("Symbole d'opération non valide : " + symbole)
        
        if not (isinstance(a, (int, float)) and isinstance(b, (int, float))):
            raise ValueError("Les valeurs d'entrée doivent être des nombres.")
        
        try:
            resultat = self.operations[symbole](a, b)
            return resultat
        except Exception as e:
            raise ValueError("Erreur lors du calcul : " + str(e))
            
# Dans le programme principal, créez une instance de la classe Calculatrice et utilisez une boucle while qui permet à 
# l'utilisateur de continuer à effectuer des calculs jusqu'à ce qu'il choisisse de quitter.

def exponentiation(a, b):
    return a ** b

def racine_carre(a, b):
    return math.sqrt(a)

def logarithme(a, b):
    if a <= 0 or b <= 0:
        raise ValueError("Les valeurs doivent être positives pour le logarithme.")
    return math.log(a, b)

calcul = calculatrice()

calcul.add_operation('^', exponentiation)
calcul.add_operation('sqrt', racine_carre)
calcul.add_operation('log', logarithme)

while True:
    print("Options disponibles:")
    print("1. Opérations de base (+, -, *, /)")
    print("2. Opérations avancées (^, sqrt, log)")
    print("3. Quitter")

# Utilisez la fonction input() pour obtenir les entrées de l'utilisateur concernant les nombres et le symbole d'opération.
# Utilisez la bibliothèque mathématique pour les opérations mathématiques avancées.
    choix = input("Choisissez une option (1/2/3) : ")

    if choix == '1':
        a = float(input("Entrez le premier nombre : "))
        symbole = input("Entrez le symbole de l'opération (+, -, *, /) : ")
        b = float(input("Entrez le deuxième nombre : "))

        try:
            resultat = calcul.calculer(a, symbole, b)
            print("Résultat :", resultat)
        except ValueError as e:
            print("Erreur :", e)

    elif choix == '2':
        a = float(input("Entrez le nombre : "))
        symbole = input("Entrez le symbole de l'opération (^, sqrt, log) : ")

        if symbole not in ('^', 'sqrt', 'log'):
            print("Opération non valide.")
            continue

        if symbole == 'log':
            base = float(input("Entrez la base du logarithme : "))
            try:
                resultat = calcul.calculer(a, symbole, base)
                print("Résultat :", resultat)
            except ValueError as e:
                print("Erreur :", e)
        else:
            b = float(input("Entrez la deuxième valeur (si nécessaire) : "))
            try:
                resultat = calcul.calculer(a, symbole, b)
                print("Résultat :", resultat)
            except ValueError as e:
                print("Erreur :", e)

    elif choix == '3':
        print("Au revoir !")
        break

    else:
        print("Option non valide. Veuillez choisir 1, 2 ou 3.")
```
